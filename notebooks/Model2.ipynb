```python
import pandas as pd
import numpy as np
```

```python
# load data
df = pd.read_csv("../data/combine.csv")
df.head()
```

```python
df = df[["Close"]]
days = 50
```

```python
col_name = f"{days}_days"
df[col_name] = df[["Close"]].shift(-days)
```

```python
df
```

```python
df_source = df.dropna()
df_source.tail()
```

```python
X = np.array(df_source[["Close"]])
```

```python
y = np.array(df_source[col_name])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
model = SVR(kernel='rbf', C=1e3, gamma=0.00001)
```

```python
model.fit(X_train, y_train)
param_grid = {'C': [1, 5, 10, 50, 1000],
              'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.00001]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
grid.score(X_test, y_test)
```
